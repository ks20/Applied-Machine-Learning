```
from PIL import Image
from sklearn.ensemble import RandomForestRegressor as RFR
from sklearn.neighbors import KNeighborsClassifier

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

#2b

```
img = np.asarray(Image.open("MonaLisa.jpg"))
plt.xlabel("Original Image")
plt.imshow(img)
plt.show()

data = list()
label = list()

num_rows = img.shape[0]
num_cols = img.shape[1]

for i in range(5000):
  data.append([np.random.randint(num_rows), np.random.randint(num_cols)])
  p_x = data[i][0]
  p_y = data[i][1]
  label.append(img[p_x, p_y])

data = np.array(data)
label = np.array(label) # True labels of the pixel coordinate

sampledImg = np.zeros([num_rows, num_cols, 3], dtype=np.uint8)
sampledImg.fill(255)

for i in range(5000):
  sampledImg[data[i][0]][data[i][1]] = label[i]

plt.xlabel("Sampled Points")
plt.imshow(sampledImg)
plt.show()
```

#2c

```
plt.xlabel("Original Image")
plt.imshow(img)
plt.show()

scaledImg = np.array(img, dtype=float)

for i in range(num_rows):
  for j in range(num_cols):
    scaledImg[i,j] = img[i,j] / 256

plt.imshow(scaledImg)
plt.xlabel("Scaled Image")
plt.show()

scaledLabel = np.array(label, dtype=float)
scaledLabel = scaledLabel / 256
```

#2d

```
def random_forest(n_tree, depth):
  rfr = RFR(random_state=0, n_estimators = n_tree, max_depth = depth)
  rfr.fit(data, label)
  pred = np.zeros([num_rows, num_cols, 3])
  
  for i in range(num_rows):
    for j in range(num_cols):
      point = np.array([i, j]).reshape(1, -1)
      pred[i, j] = rfr.predict(point) / 256

  return pred
```

```
pred = random_forest(1, None)
plt.xlabel("Predicted")
plt.imshow(pred)
plt.show()
```

```
def knn():
  knn = KNeighborsClassifier(n_neighbors=1)
  knn.fit(data, label)
  pred = np.zeros([num_rows, num_cols, 3])
  
  for i in range(num_rows):
    for j in range(num_cols):
      point = np.array([i, j]).reshape(1, -1)
      pred[i, j] = knn.predict(point) / 255
    
  return pred
```

```
pred = knn()
plt.xlabel("KNN")
plt.imshow(pred)
plt.show()
```
